```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import scienceplots


#plt.style.use(['science','nature'])
plt.style.use('sci.mplstyle')
```

```python
folder = "Results"
suffix = ""
```

```python
target_losses = np.load(f"{folder}/target_losses.npy")
target_losses_val = np.load(f"{folder}/target_losses_val.npy")

control_losses = np.load(f"{folder}/control_losses.npy")
control_losses_val = np.load(f"{folder}/control_losses_val.npy")

target_fidelity = np.load(f"{folder}/target_fidelity.npy")
target_fidelity_val = np.load(f"{folder}/target_fidelity_val.npy")

control_fidelity = np.load(f"{folder}/control_fidelity.npy")
control_fidelity_val = np.load(f"{folder}/control_fidelity_val.npy")
```

```python
bernardi_sqp_mean = np.load(f"{folder}/bernardi_sqp_mean.npy")
proposed_sqp_mean = np.load(f"{folder}/proposed_sqp_mean.npy")
exact_sqp_mean = np.load(f"{folder}/exact_sqp_mean.npy")

bernardi_sqp_error = np.load(f"{folder}/bernardi_sqp_error.npy")
proposed_sqp_error = np.load(f"{folder}/proposed_sqp_error.npy")

bernardi_fidelity_mean = np.load(f"{folder}/bernardi_fidelity_mean.npy")
proposed_fidelity_mean = np.load(f"{folder}/proposed_fidelity_mean.npy")
```

```python
bernardi_time_mean = np.load(f"{folder}/bernardi_time_mean.npy")
proposed_time_mean = np.load(f"{folder}/proposed_time_mean.npy")
exact_time_mean = np.load(f"{folder}/exact_time_mean.npy")
```

```python
semiprimes = np.load(f"{folder}/semiprimes.npy")
a_probs = np.load(f"{folder}/a-probabilities.npy")

probs = {}

for i in range(semiprimes.shape[0]):
    probs[semiprimes[i]] = a_probs[i]

probs = dict(sorted(probs.items(), key=lambda item: item[0]))

print(probs)

keys = np.fromiter(probs.keys(), dtype=int)
vals = np.fromiter(probs.values(), dtype=float)

vals_norm = keys * vals
```

```python
folder = "Results"
suffix = ""
```

```python
fig = plt.figure(figsize=(5, 8))

ax1a = fig.add_axes([0.00, 0.70, 1.00, 0.30])
ax1b = fig.add_axes([0.00, 0.35, 1.00, 0.30])
ax1c = fig.add_axes([0.00, 0.00, 1.00, 0.30])


# Add the x and y-axis labels
ax1a.set_ylabel(r'$\mathcal{E}_{\mathrm{SQP}}$'    , labelpad=1.0)
ax1b.set_ylabel(r'$\bar{F}$'                       , labelpad=1.0)
ax1c.set_ylabel(r'Mean SQP'                        , labelpad=1.0)
ax1c.set_xlabel(r'Step'                            , labelpad=1.5)

# Set the axis limits
ax1a.set_ylim(-0.1, 1)
ax1b.set_ylim(-0.1, 1)
ax1c.set_ylim(-0.1, 1)


# Plot Figure 3a
ax1a.text(-30.0, 1, '(a)', fontweight='bold')
ax1a.axhline(0, 0, bernardi_sqp_error.shape[0], color='gray', linewidth=1)
ax1a.plot(np.arange(bernardi_sqp_error.shape[0]), bernardi_sqp_error, color="r", label="Bernardi", linestyle="dashed")
ax1a.plot(np.arange(proposed_sqp_error.shape[0]), proposed_sqp_error, color="g", label="Proposed", linestyle="solid")
ax1a.plot(np.arange(proposed_sqp_error.shape[0]), bernardi_sqp_error - proposed_sqp_error, color="b", label="Bernardi - Proposed", linestyle="dashdot")


# Plot Figure 3b
ax1b.text(-30.0, 1.0, '(b)', fontweight='bold')
ax1b.axhline(0, 0, bernardi_sqp_error.shape[0], color='gray', linewidth=1)
ax1b.plot(np.arange(bernardi_sqp_error.shape[0]), bernardi_fidelity_mean, color="r", label="Bernardi", linestyle="dashed")
ax1b.plot(np.arange(proposed_sqp_error.shape[0]), proposed_fidelity_mean, color="g", label="Proposed", linestyle="solid")
ax1b.plot(np.arange(proposed_sqp_error.shape[0]), proposed_fidelity_mean - bernardi_fidelity_mean, color="b", label="Proposed - Bernardi", linestyle="dashdot")


# Plot Figure 3c
ax1c.text(-30.0, 1, '(c)', fontweight='bold')
ax1c.axhline(0.5, 0, bernardi_sqp_error.shape[0])
ax1c.plot(np.arange(bernardi_sqp_error.shape[0]), bernardi_sqp_mean, color="r", label="Bernardi", linestyle="dashed")
ax1c.plot(np.arange(proposed_sqp_error.shape[0]), proposed_sqp_mean, color="g", label="Proposed", linestyle="solid")

leg = ax1a.legend(loc='upper right', fontsize=12, fancybox=True, handlelength=1.5)
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())

leg = ax1b.legend(loc='right', fontsize=12, fancybox=True, handlelength=1.5)
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())

leg = ax1c.legend(loc='upper right', fontsize=12, fancybox=True, handlelength=1.5)
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())

fig.show()

fig.savefig(f"{folder}/Figure1{suffix}.pdf", bbox_inches="tight")
```

```python
fig = plt.figure(figsize=(5, 8))

ax2a = fig.add_axes([0.00, 0.50, 1.00, 0.45])
ax2b = fig.add_axes([0.00, 0.00, 1.00, 0.45])


# Add the x and y-axis labels
ax2a.set_ylabel(r'$\#$ of $\mathscr{ɑ}_s$'                         , labelpad=1.0)
ax2b.set_ylabel(r'$\#$ of $\mathscr{ɑ}_s / N_s$'                  , labelpad=1.0)
ax2b.set_xlabel(r'$N_s$'                           , labelpad=1.5)

# Set the axis limits

ax2a.set_yscale("log")
ax2b.set_yscale("log")

# Plot Figure 3a
ax2a.text(-8000, 10000, '(a)', fontweight='bold')
ax2a.scatter(keys, vals_norm, label=r"$\#$ of $\mathscr{ɑ}_s$")


# Plot Figure 3b
ax2b.text(-8000, 1, '(b)', fontweight='bold')
ax2b.scatter(keys, vals, label=r"$\#$ of $\mathscr{ɑ}_s / N_s$")

fig.show()

fig.savefig(f"{folder}/Figure2{suffix}.pdf", bbox_inches="tight")
```

```python
fig = plt.figure(figsize=(16, 4))

ax3a = fig.add_axes([0.00, 0.00, 0.40, 1.00])
ax3b = fig.add_axes([0.50, 0.00, 0.40, 1.00])


# Add the x and y-axis labels
ax3a.set_ylabel(r'$\mathcal{E}_{\mathrm{RMS1F}}$'  , labelpad=1.0)
ax3b.set_ylabel(r'Mean Fidelity'                   , labelpad=1.0)
ax3a.set_xlabel(r'Epoch'                           , labelpad=1.5)
ax3b.set_xlabel(r'Epoch'                           , labelpad=1.5)

# Set the axis limits
ax3a.set_ylim(0, 0.2)
ax3b.set_ylim(0.8, 1)

# Plot Figure 3a
ax3a.text(-150, 0.2, '(a)', fontweight='bold')
ax3a.plot(np.arange(target_losses.shape[0]), target_losses, label="Target NN Training Loss", linestyle="solid")
ax3a.plot(np.arange(target_losses.shape[0]), target_losses_val, label="Target NN Validation Loss", linestyle="dashed")

ax3a.plot(np.arange(target_losses.shape[0]), control_losses, label="Control NN Training Loss", linestyle="dotted")
ax3a.plot(np.arange(target_losses.shape[0]), control_losses_val, label="Control NN Validation Loss", linestyle="dashdot")


# Plot Figure 3b
ax3b.text(-150, 1, '(b)', fontweight='bold')
ax3b.plot(np.arange(target_fidelity.shape[0]), target_fidelity, label="Target NN Training Fidelity", linestyle="solid")
ax3b.plot(np.arange(target_fidelity.shape[0]), target_fidelity_val, label="Target NN Validation Fidelity", linestyle="dashed")

ax3b.plot(np.arange(target_fidelity.shape[0]), control_fidelity, label="Control NN Training Fidelity", linestyle="dotted")
ax3b.plot(np.arange(target_fidelity.shape[0]), control_fidelity_val, label="Control NN Validation Fidelity", linestyle="dashdot")


leg = ax3a.legend(loc='right', fontsize=12, fancybox=True, handlelength=1.5)
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())

leg = ax3b.legend(loc='right', fontsize=12, fancybox=True, handlelength=1.5)
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())

fig.show()

fig.savefig(f"{folder}/Figure3{suffix}.pdf", bbox_inches="tight")
```

```python
fig = plt.figure(figsize=(8, 4))

ax4a = fig.add_axes([0.00, 0.00, 0.40, 1.00])
ax4b = fig.add_axes([0.50, 0.00, 0.40, 1.00])


# Add the x and y-axis labels
ax4a.set_ylabel(r'Time (s)'                        , labelpad=1.0)
ax4a.set_xlabel(r'Number of Qubits'                , labelpad=1.5)
ax4b.set_xlabel(r'Number of Qubits'                , labelpad=1.5)

ax4a.set_yscale('log')

# Plot Figure 4a
plt.figtext(-0.13, 0.97, '(a)', fontweight='bold')
ax4a.plot(np.arange(proposed_time_mean.shape[0]) + 2, exact_time_mean, color="b", label="Exact", linestyle="dashdot", marker="D", ms=10)
ax4a.plot(np.arange(bernardi_time_mean.shape[0]) + 2, bernardi_time_mean, color="r", label="Bernardi", linestyle="dashed", marker="s", ms=10)
ax4a.plot(np.arange(proposed_time_mean.shape[0]) + 2, proposed_time_mean, color="g", label="Proposed", linestyle="solid", marker="o", ms=10)

ax4a.set_xticks(range(2, 13, 2))
ax4a.set_xticks(range(3, 13, 2), minor=True)
ax4a.set_yticks(np.logspace(-4, 2, 4))


# Plot Figure 4b
plt.figtext(0.41, 0.97, '(b)', fontweight='bold')
ax4b.plot(np.arange(proposed_time_mean.shape[0]) + 2, exact_time_mean, color="b", label="Exact", linestyle="dashdot", marker="D", ms=10)
ax4b.plot(np.arange(bernardi_time_mean.shape[0]) + 2, bernardi_time_mean, color="r", label="Bernardi", linestyle="dashed", marker="s", ms=10)
ax4b.plot(np.arange(proposed_time_mean.shape[0]) + 2, proposed_time_mean, color="g", label="Proposed", linestyle="solid", marker="o", ms=10)

ax4b.set_xticks(range(2, 13, 2))
ax4b.set_xticks(range(3, 13, 2), minor=True)

leg = ax4b.legend(loc='upper left', fancybox=True, handlelength=1.5)
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())


fig.show()

fig.savefig(f"{folder}/Figure4{suffix}.pdf", bbox_inches="tight")
```
